# Encoding

```python
encoder = OneHotEncoder()
```

```python
train['Over18'] = train['Over18'].map({'Y': 1, 'N': 0})
train['OverTime'] = train['OverTime'].map({'No': 0, 'Yes': 1})
train['Gender'] = train['Gender'].map({'Male': 1, 'Female': 0})
```

```python
# encoder.fit(train['Travel_Rarely'].values.reshape(-1, 1))
```

```python
f_c = RandomForestClassifier()
```

```python
X = train.drop(['Attrition', 'BusinessTravel', 'Gender',  'Department', 'EducationField', 'JobRole', 'MaritalStatus'], axis=1)
y = train['Attrition']
```

```python
X.head()
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
```

```python
# X_train.columns
```

# Fit to model

```python
f_c.fit(X_train, y_train)
```

```python
y_prediction = f_c.predict(X_val)
```

```python
accuracy = accuracy_score(y_val, y_prediction)
```
